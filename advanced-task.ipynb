```python
from transformers import DetrImageProcessor, DetrForObjectDetection
import torch
from PIL import Image
processor_dino = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50")
model_dino = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50")
```

```python
from transformers import AutoProcessor, AutoModelForVision2Seq
import torch
from PIL import Image

model_id = "llava-hf/llava-v1.6-mistral-7b-hf"

processor = AutoProcessor.from_pretrained(model_id)
model = AutoModelForVision2Seq.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

```

```python
image_path = "/kaggle/input/dataset/fruits.jpg"
image = Image.open(image_path)
import json



inputs = processor_dino(images=image, return_tensors="pt")

outputs = model_dino(**inputs)

target_sizes = torch.tensor([image.size[::-1]])  # (height, width)
results = processor_dino.post_process_object_detection(outputs, target_sizes=target_sizes, threshold=0.5)[0]

objects_dict = {}

for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
    name = model_dino.config.id2label[label.item()]
    bbox = [round(x.item(), 2) for x in box]
    conf = round(score.item(), 3)
    if conf > 0.5 :
        entry = {"bbox": bbox, "confidence": conf}
    
        if name in objects_dict:
            if isinstance(objects_dict[name], list):
                objects_dict[name].append(entry)
            else:
                objects_dict[name] = [objects_dict[name], entry]
        else:
            objects_dict[name] = entry

json_string = json.dumps(objects_dict, indent=4)
print(json_string)
```

```python
image = Image.open(image_path).convert("RGB") 
# image_input = preprocess(image).unsqueeze(0).to(device)
```

```python
prompt = f"""
<image>
You are an ESCA-style scene graph generator.

Extract a scene graph from this image with the following constraints:

1. Identify all objects with unique IDs ("obj1", "obj2", ...).
2. Include attributes for every object (color, size, material, state).
3. Include pairwise relationships between objects.
4. Valid predicates you can use are ("inside","on", "next to", "behind" , "in front of").
6. Output ONLY valid JSON.
7. Do NOT add explanation|s or natural language outside of JSON.
8. JSON structure must be:

{{
  "objects": {{
      "obj1": {{"name": "...", "attributes": ["...", "..."]}}
  }},
  "relationships": [
      {{"subject": "obj1", "predicate": "....", "object": "obj2", }},
      {{"subject": "obj1", "predicate": "....", "object": "obj3", }}
  ]
}}

Here are the detected objects from DINO with bounding boxes and confidences:
{json_string}

Use the DINO objects as ground truth. 
Return ONLY the JSON.
"""
```

```python
inputs = processor(
    text=prompt,
    images=image,
    return_tensors="pt"
).to("cuda", torch.float16)

with torch.no_grad():
    output_ids = model.generate(
        **inputs,
        max_new_tokens=600,
        do_sample=True,
        temperature=0.1
    )

raw_output = processor.decode(output_ids[0], skip_special_tokens=True)
print(raw_output)
```

```python
text = raw_output
word = "Return ONLY the JSON."

index = text.find(word)
if index != -1:
    new_text = text[index + len(word):]  # start after the word"
else:
    new_text = text  # word not found

print(new_text.strip())  # strip() removes leading spaces

```

```python

import json,re
def extract_json(text):
    # Find the first {...} block
    match = re.search(r"\{.*\}", text, re.DOTALL)
    if not match:
        raise ValueError("No JSON object found in output.")

    json_str = match.group(0)

    # Try raw load
    try:
        return json.loads(json_str)
    except json.JSONDecodeError:
        pass

    # Auto-fix common issues
    json_str = json_str.replace("\n", "")
    json_str = re.sub(r",\s*}", "}", json_str)
    json_str = re.sub(r",\s*]", "]", json_str)

    return json.loads(json_str)
```

```python
scene_graph = extract_json(new_text)
print(json.dumps(scene_graph, indent=2))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

model_id = "mistralai/Mistral-7B-Instruct-v0.3"

# Configure 4-bit quantization to fit on consumer GPUs (T4/L4/A10)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

print("Loading Mistral tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_id)

print("Loading Mistral model...")
llm_model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

print("Mistral loaded successfully.")
```

```python
# You can change this question to anything relevant to the image/scene graph
user_query = "Is coconut above the apple"
# user_query = "Is there any food mentioned in the scene?"

print(f"User Query: {user_query}")
```

```python
import json

# Ensure scene_graph is a string for the prompt
scene_graph_str = json.dumps(scene_graph, indent=2)

# Construct the Prompt using Mistral's instruction format [INST] ... [/INST]
prompt = f"""
[INST]
You are an intelligent reasoning assistant.  
You answer questions ONLY using the information found in a Scene Graph (JSON).  
You must reason step-by-step using the structure of the JSON and then produce a short final answer.  
If the answer cannot be found in the JSON, say so explicitly.

Below are **examples** of how you should think and answer.

====================  
### EXAMPLE 1  
--- SCENE GRAPH ---
{{
  "objects": [
    {{ "id": 1, "name": "dog", "attributes": ["brown"], "bbox": [10, 20, 80, 60] }},
    {{ "id": 2, "name": "ball", "attributes": ["red"], "bbox": [85, 40, 100, 55] }}
  ],
  "relations": [
    {{ "subject": 1, "predicate": "next to", "object": 2 }}
  ]
}}
-------------------
USER QUESTION: What object is next to the dog?

### REASONING
1. Look for object with name "dog" → id = 1.  
2. Check relations where subject = 1.  
3. There is a relation: dog (1) "next to" object 2.  
4. Object 2 has name "ball".  

### FINAL ANSWER  
The ball is next to the dog.

====================  
### EXAMPLE 2  
--- SCENE GRAPH ---
{{
  "objects": [
    {{ "id": "A", "name": "car", "attributes": ["blue"], "bbox": [0,0,50,40] }},
    {{ "id": "B", "name": "tree", "attributes": ["tall"], "bbox": [60,0,80,100] }}
  ],
  "relations": [
    {{ "subject": "A", "predicate": "in front of", "object": "B" }}
  ]
}}
-------------------
USER QUESTION: What is in front of the tree?

### REASONING
1. Find object named "tree" → id = B.  
2. Look for relations where object = B and predicate = "in front of".  
3. Relation: A "in front of" B.  
4. Object A is a "car".  

### FINAL ANSWER  
The car is in front of the tree.

====================  
### EXAMPLE 4 (Transitivity Example)

--- SCENE GRAPH ---

{{
  "objects": [
    {{ "id": "T", "name": "table", "attributes": ["wooden"], "bbox": [0, 50, 200, 100] }},
    {{ "id": "B", "name": "book",  "attributes": ["red"],    "bbox": [40, 30, 120, 50] }},
    {{ "id": "P", "name": "pen",   "attributes": ["blue"],   "bbox": [60, 10, 100, 30] }}
  ],
  "relations": [
    {{ "subject": "B", "predicate": "on top of", "object": "T" }},
    {{ "subject": "P", "predicate": "on top of", "object": "B" }}
  ]
}}

USER QUESTION: What object is above the table?

REASONING

The table is object T.

Check all relations that end with object = T.

Relation: B (book) → on top of → T.
So the book is directly above the table.

Next, check if any object is on top of the book.

Relation: P (pen) → on top of → B.
So the pen is above the book, which is above the table.

Therefore, both B (book) and P (pen) are above the table by transitivity.

FINAL ANSWER

The book and the pen are above the table.

### NOW YOUR TASK  
Use the same style: step-by-step reasoning using ONLY the provided Scene Graph, then a short final answer.

--- SCENE GRAPH ---
{scene_graph_str}
-------------------

USER QUESTION: {user_query}

### REASONING
[/INST]
"""



# Tokenize and Generate
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

with torch.no_grad():
    outputs = llm_model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.1, # Low temperature for factual reasoning
        pad_token_id=tokenizer.eos_token_id
    )

# Decode the output
full_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Extract only the assistant's new text (removing the prompt)
answer = full_response.split("[/INST]")[-1].strip()

print("="*40)
print(f"QUERY: {user_query}")
print("="*40)
print("MISTRAL ANSWER:")
print(answer)
```
